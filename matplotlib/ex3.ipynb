```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-label')
plt.ylabel('Y-label')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=40)
plt.ylabel('Y-Axis', labelpad=20)

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])

plt.xlabel('X-Axis', labelpad=15, fontdict={ 'family': 'serif', 'color': 'b', 'weight': 'bold', 'size': 14 })
plt.ylabel('Y-Axis', labelpad=20, fontdict={ 'family': 'fantasy', 'color': 'dee[pink', 'weight': 'normal', 'size': 'xx-large' })

plt.show()
```

```python
font1 = { 'family': 'serif',
          'color': 'b',
          'weight': 'bold',
          'size': 14 
        }


font2 = { 'family': 'fantasy',
          'color': 'deeppink',
          'weight': 'normal',
          'size': 'xx-large'
        }

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15, fontdict=font1)
plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', loc='left')
plt.ylabel('Y-Axis', loc='top')
plt.show()
```
